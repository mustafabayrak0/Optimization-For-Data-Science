Import necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

Define the function

```python
def f(x1, x2):
    return ((x1 ** 4) / 12) - ((x1 ** 2) / 2) + ((x2 ** 3) / 6) + (x2 ** 2) + ((3 * x2) / 2) + 20
```

Create a meshgrid of x1 and x2 values. To contain all stationary points, I have chose boundaries as -4 and 4.

```python
x1_values = np.linspace(-4, 4, 100)
x2_values = np.linspace(-4, 4, 100)
x1_mesh, x2_mesh = np.meshgrid(x1_values, x2_values)
```

Compute the values of function

```python
vals = f(x1_mesh, x2_mesh)
```

**(c) Plot 3D surface in any style in the appropriate interval containing all stationary points**

```python
fig = plt.figure(figsize=(5,8))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x1_mesh, x2_mesh, vals, cmap='viridis')
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('f(x1,x2)')
plt.show()
```

**(d) Plot contour plot and show (overlay) the gradient directions**

Compute the gradient of the function and normalize

```python
x1_grad = (x1_mesh**3)/3 - x1_mesh
x2_grad = (x2_mesh**2)/2 + 2*x2_mesh + 3/2
grad_norm = np.linalg.norm(np.array([x1_grad, x2_grad]), axis=0)
grad_norm_x1,grad_norm_x2 = x1_grad / grad_norm, x2_grad / grad_norm
```

Plot contour plot and show (overlay) the gradient directions

```python
fig, ax = plt.subplots()
contour = ax.contour(x1_mesh, x2_mesh, vals, levels=20)
ax.quiver(x1_mesh, x2_mesh, grad_norm_x1, grad_norm_x2)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.clabel(contour)
plt.show()
```
